```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("german_credit_data.csv")
```

```python
# To see the distribution of Gender
sns.countplot(x='Sex', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
# To see the distribution of Housing status
sns.countplot(x='Housing', data=df)
plt.title("Housing Distribution")
plt.show()
```

```python
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[18,30,40,50,60,100],
    labels=['18-30','31-40','41-50','51-60','60+']
)
```

```python
sns.countplot(
    x='Age_Group',
    hue='Sex',
    data=df
)
plt.title("Age Group vs Gender")
plt.show()
```

```python
sns.boxplot(
    x='Housing',
    y='Duration',
    data=df
)
plt.title("Loan Duration by Housing Status")
plt.show()
```

```python
sns.boxplot(
    x='Housing',
    y='Credit amount',
    data=df
)
plt.title("Credit Amount by Housing Status")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,8))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
# Add derived columns for deeper analysis
df['Age_Group'] = pd.cut(df['Age'],
    bins=[0,25,35,50,75],
    labels=['18-25','26-35','36-50','51+'])

df['Credit_Band'] = pd.cut(df['Credit amount'],
    bins=[0,1000,3000,6000,10000,25000],
    labels=['<1K','1K-3K','3K-6K','6K-10K','>10K'])

# Fill missing values (183 missing saving accounts, 394 checking accounts)
df['Saving accounts'] = df['Saving accounts'].fillna('unknown')
df['Checking account'] = df['Checking account'].fillna('unknown')

print("Cleaning done. Shape:", df.shape)
```

```python
# Which loan purpose is riskiest?
# Since dataset has no 'Risk' column, we use Credit amount > 5000 AND Duration > 30 as a proxy for high-risk
df['High_Risk_Proxy'] = ((df['Credit amount'] > 5000) & (df['Duration'] > 30)).astype(int)

risk_by_purpose = df.groupby('Purpose')['High_Risk_Proxy'].mean().sort_values(ascending=False)
print("High-risk rate by loan purpose:")
print((risk_by_purpose * 100).round(1))

plt.figure(figsize=(10,5))
risk_by_purpose.plot(kind='bar', color='#E24B4A')
plt.title("High-Risk Loan Rate by Purpose")
plt.ylabel("Risk Rate")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('risk_by_purpose.png', dpi=150)
plt.show()
```

```python
plt.figure(figsize=(9,5))
colors = df['Housing'].map({'own':'#1D9E75','free':'#378ADD','rent':'#E24B4A'})
plt.scatter(df['Credit amount'], df['Duration'], c=colors, alpha=0.5, s=20)
plt.xlabel("Credit Amount")
plt.ylabel("Duration (months)")
plt.title("Credit Amount vs Duration (color = Housing)")
plt.tight_layout()
plt.savefig('scatter_amount_duration.png', dpi=150)
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.boxplot(x='Saving accounts', y='Credit amount', data=df,
            order=['unknown','little','moderate','quite rich','rich'],
            palette='Blues')
plt.title("Credit Amount by Savings Level")
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig('savings_vs_credit.png', dpi=150)
plt.show()
```
